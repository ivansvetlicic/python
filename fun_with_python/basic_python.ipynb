```
#given a list of whole numbers, find 2 numbers whose sum is equal to the target number
#e.g. [2, 4, 3, 1] if target = 3, print 2 and 1
lst = [2, 4, 3, 1]
target = 3
solutions = []
for i in range(len(lst)):
  for j in range(len(lst)):
    if lst[i] + lst[j] == target:
      solutions.append((lst[i], lst[j]))

print(solutions)
```

```
#let's optimze the solution above
#given a list of whole numbers, find 2 numbers whose sum is equal to the target number
#e.g. [2, 4, 3, 1] if target = 3, print 2 and 1
lst = [2, 2, 1]
target = 4
def find_pairs(lst, target):
  solutions = []
  i = 1
  for elem1 in lst:
    for elem2 in lst[i :]:
      if elem1 + elem2 == target:
        solutions.append((elem1, elem2))
    i += 1
    print("for list: ", lst, "and target:", target, ", solutions are:", solutions)
    return solutions
def run_tests(test_cases):
  for test in test_cases:
    find_pairs(test[0], test[1])

#define test cases:
test_cases = []
test_cases.append([[2, 2, 1], 4])
test_cases.append([[1, 2, 3, 2, 0, 7], 4])
test_cases.append([[1, 1, 1], 1])
run_tests(test_cases)
```

```
#string comparisons
def words_matching (s1, s2):
  if s1.upper() == s2.upper():
    return True
  else:
    return False

def words_matching_1 (s1, s2):
  s1 = s1.upper()
  s2 = s2.upper()

  if len(s1) != len(s2):
    return False
  
  if hash(s1) != hash(s2):
    return False

  for i in range(len(s1)):
    if s1[i] != s2[i]:
      return False

  return True

words_matching_1('opp kj', 'OpP kj')
```

```
def do_reverse(lst):
  reversed = []
  for char in lst:
    reversed.insert(0, char)
  return ''.join(reversed)

lst = 'we want this in reverse'

print(do_reverse(lst))
```

```
#given two lists, find a missing element from the 1st list (1st list will always have a missing element)
lst1 = [3,2,7,9]
lst2 = [3,2,7,5,9]

def find_longer(lst1, lst2):
  if len(lst1) > len(lst2):
    return lst1, lst2
  else:
    return lst2, lst1

def find_missing(lst1, lst2):
  lst1, lst2 = find_longer(lst1, lst2)
  for elem in lst1:
    if elem not in lst2:
      print ('missing element:', elem)

def find_missing_v2(lst1, lst2):
  lst1, lst2 = find_longer(lst1, lst2)
  lst1.sort()
  lst2.sort()
  for elem1, elem2 in zip(lst1, lst2):
    if elem1 != elem2:
      print ('missing element:', elem1)
      return

find_missing_v2(lst1, lst2)
```
